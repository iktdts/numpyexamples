Solar Storm

```python
from tkinter import Tk, filedialog

Tk().withdraw() 
filename = filedialog.askopenfilenames() 
```

```python
import numpy as np
import cv2 as cv

KERNEL_SIZE=11
IMG_SD=5
THRESHOLD=180
KERNEL_SIZE_MASK=51
IMG_SD_MASK=21


img=cv.imread(filename[0])
bw_img=img.copy()
bw_img = cv.cvtColor(bw_img, cv.COLOR_BGR2GRAY)

hist, bin_edges = np.histogram(bw_img, bins=20)
bin_centers = 0.5*(bin_edges[:-1] + bin_edges[1:])

bw_img = cv.GaussianBlur(bw_img,(KERNEL_SIZE,KERNEL_SIZE),IMG_SD)
binary_img = (bw_img > THRESHOLD).astype('float')
mask=cv.GaussianBlur(binary_img,(KERNEL_SIZE_MASK,KERNEL_SIZE_MASK),IMG_SD)
mask=(mask>0).astype(np.uint)
final_img=img*np.repeat(mask[..., np.newaxis], 3, axis=2)
cv.imwrite('images/storms.png', final_img)

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,20))
plt.subplot(231)
plt.imshow(img[...,[2,1,0]])
plt.axis('off')
plt.subplot(232)
plt.imshow(255-bw_img,cmap='Greys')
plt.axis('off')
plt.subplot(233)
plt.subplot(233).title.set_text("Histogram")
plt.plot(bin_centers[1:], hist[1:], lw=2)
plt.axvline(THRESHOLD, color='r', ls='--', lw=2)
plt.yticks([])
plt.subplot(234)
plt.imshow(binary_img, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(235)
plt.imshow(mask, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(236)
plt.imshow(final_img[...,[2,1,0]])
plt.axis('off')
plt.subplots_adjust(wspace=0.1, hspace=0)
plt.savefig("images/process.png")
plt.show()
```
